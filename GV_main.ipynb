```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt
data=pd.read_csv("Gender_voice.csv")
data.info()
```

```python
data["label"]=data["label"].replace({"male":0,"female":1}) #replacing  values in the column for sex rappresentation in int
```

```python
data.head()
```

```python
from scipy.stats import zscore
dataz=data.copy()
for col in list(dataz.columns):
    dataz[col] = zscore(dataz[col])
dataz
```

```python
data['label'].value_counts().sort_index().plot.bar(figsize=(12,6))
plt.grid()
plt.show()
```

```python
bins_struges=int(3.3*np.log(len(data['meanfreq'])))
bins_rice=int(2*len(data['meanfreq'])**(1/3))
plt.subplot(2,2,1)
plt.title('Struges ({} bins)'.format(bins_struges))
data['meanfreq'].plot.hist(figsize=(12,6), density='True', bins=bins_struges)
plt.grid()
plt.subplot(2,2,2)
plt.title('Rice ({} bins)'.format(bins_rice))
data['meanfreq'].plot.hist(figsize=(12,6), density='True', bins=bins_rice)
plt.grid()
plt.subplot(2,2,3)
data['IQR'].plot.hist(figsize=(12,6), density='True', bins=bins_struges)
plt.grid()
plt.subplot(2,2,4)
data['IQR'].plot.hist(figsize=(12,6), density='True', bins=bins_rice)
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.title('Struges ({} bins)'.format(bins_struges))
data['IQR'].plot.hist(figsize=(12,6), density='True', bins=bins_struges)
plt.grid()
plt.subplot(1,2,2)
plt.title('Rice ({} bins)'.format(bins_rice))
data['IQR'].plot.hist(figsize=(12,6), density='True', bins=bins_rice)
plt.grid()
plt.show()
```

```python
dataz.boxplot(by='label',column=['meanfreq',"IQR"], figsize=(8,6))
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.subplot(121)
plt.title('Mean Freq')
dataz['meanfreq'].plot.density()
plt.subplot(122)
plt.title('IQR')
dataz['IQR'].plot.density()
plt.show()
```

```python
plt.figure(figsize=(16,10))
plt.subplot(121)
a=dataz['meanfreq'].plot.density(figsize=(12,6))
plt.hist(dataz['meanfreq'], density=True)
plt.grid()
plt.subplot(122)
b=dataz['IQR'].plot.density(figsize=(12,6))
plt.hist(dataz['IQR'], density=True)
plt.grid()
plt.show()
```

```python
a=dataz[dataz['label']==-1]['meanfreq'].value_counts().sort_index().cumsum()
b=dataz[dataz['label']==1]['meanfreq'].value_counts().sort_index().cumsum()
plt.figure(figsize=(16,8))
plt.subplot(121)
plt.plot(a.index,a.values)
plt.plot(b.index,b.values)
plt.grid()
plt.legend(['F','M'])
c=dataz[dataz['label']==-1]['IQR'].value_counts().sort_index().cumsum()
d=dataz[dataz['label']==1]['meanfreq'].value_counts().sort_index().cumsum()
plt.subplot(122)
plt.plot(c.index,c.values)
plt.plot(d.index,d.values)
plt.grid()
plt.legend(['F','M'])
plt.show()
plt.show()
```

```python
dataz.groupby('label')['meanfreq'].plot.hist(width=0.4, alpha=0.8, density=True, figsize=(16,6))
dataz.groupby('label')['meanfreq'].plot.density()
plt.legend(['F','M','F','M'])
plt.grid()
plt.show()
```
